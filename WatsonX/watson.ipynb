```python
from ibm_watsonx_ai import APIClient, Credentials
from ibm_watsonx_ai.foundation_models import ModelInference

credentials = Credentials(
    url=f"{watsonx_ai_url}",
    api_key=f"{apikey}",
)
client = APIClient(credentials)

model = ModelInference(              # ✅ 들여쓰기 제거
    model_id="ibm/granite-4-h-small",
    api_client=client,
    project_id=f"{project_id}",      # ✅ f-string 수정
    params={
        "max_tokens": 100
    }
)

#사용자 프롬프트
prompt = 'How far is Paris from Bangalore?'


#시스템 프롬프트(선택적인 것. AI가 어떤 방식으로 하면 좋을 지 설정하는 것)
messages = [
    {"role":"system", "content":"You are a helpful assistant."},
    {"role":"user", "content":prompt},
]


#생성되는 자료(답변)

generated_response = model.chat(messages=messages)


print(generated_response)
print(generated_response['choices'][0]['message']['content'])
```

```python
from ibm_watsonx_ai import APIClient, Credentials
from ibm_watsonx_ai.foundation_models import ModelInference

credentials = Credentials(
    url=f"{watsonx_ai_url}",
    api_key=f"{apikey}",
)
client = APIClient(credentials)

model = ModelInference(
    model_id="ibm/granite-4-h-small",
    api_client=client,
    project_id=f"{project_id}",
    params={
        "max_new_tokens": 1000,   # ✅ 쉼표 추가 + 파라미터명 수정
        "temperature": 1
    }
)

# 사용자 프롬프트
user_prompt = """무서운 이야기 해봐."""         # ✅ 변수명 수정

# 시스템 프롬프트
system_prompt = """넌 유명 공포 소설 작가야."""  # ✅ 변수명 수정

messages = [
    {"role": "system", "content": system_prompt},   # ✅ 따옴표 제거
    {"role": "user",   "content": user_prompt},     # ✅ 따옴표 제거
]

generated_response = model.chat(messages=messages)

print(generated_response)
print(generated_response['choices'][0]['message']['content'])
```
