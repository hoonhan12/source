`````python
def add_many(choice, *args):
    if choice == "mul":
         result = 1
        for i in args:
        result *= i
    elif choice == "add":
        result = 0
        for i in args:
        result += i
    return result

print(add_many("add", 1,2,3,4,5))
print(add_many("mul", 1,2,3,4,5))
`````

`````python
# info.txt 읽어서 bmi 지수 계산
# 아하 94 188
# 이름 : 아하
# 몸무게: 94
# 키 : 154
# BMI : (weight / height/100 ** 2)
# 결과 : 
`````

`````python
with open("info.txt", "r", encoding="utf-8") as f:
    for info in f:
        name, weight, height = info.strip().split(", ")

        bmi = int(weight) / (int(height) / 100) ** 2

        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"

        print("\n".join(["이름 : {}", "몸무게 : {}", "키 : {}", "BMI : {:.2f}", "결과 : {}"]).format(name, weight, height, bmi, result))
        print()
`````

`````python
# 클래스 생성
class calculator: 
    def __init__(self):
        self.result = 0
    # 매서드
    def add(self, num):
        self.result += num
        return self.result
    def sub(self, num):
        self.result -= num
        return self.result

`````

`````python
# 사용 - 객체 생성
calc = calculator()

# 함수(메소드) 호출

print(calc.add(5))

print(calc.sub(3))
`````

`````python
class Fourcalc:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def sub(self):
        return self.first - self.second

    def mul(self):
        return self.first * self.second

    def div(self):              # ✅ 공백 4칸으로 통일
        return self.first // self.second
`````

`````python
class FourCal:          # ✅ FourCal로 통일
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def sub(self):
        return self.first - self.second

    def mul(self):
        return self.first * self.second

    def div(self):
        return self.first // self.second
    def __str__(self):
        return f"{self.first}, {self.second}"
    

# 사용
calc1 = FourCal(5, 3)
print(calc1.add())   # 8
print(calc1.sub())   # 2
print(calc1.mul())   # 15
print(calc1.div())   # 1
print(calc1)   # 1
`````

`````python
# 상속

class MoreFourCal(FourCal):
    def pow(self):
        return self.first ** self.second
`````

`````python
more = MoreFourCal(4,2)

print(more.pow())
`````

`````python
# 오버라이팅: 부모 클래스 메소드 재정의 (오류 방지하기 위해서)

four1 = FourCal(4,0)
# four1.div() ZeroDivisionError
`````

`````python
# super() : 부모
# self : 자기자신

class SafeFourCal(FourCal):
    # 오버라이딩 : 자식에서 재정의
    def div(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
`````

`````python
four1 = SafeFourCal(4,0)
four1.div()
`````

`````python
# 나만의 오류 작성하기

class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)
    
`````

`````python
import json

with open("test.json", "r") as f:
    data = json.load(f)
    print(type(data))          # <class 'list'>
    print(data)                # 전체 데이터 출력

    print(data[0]["age"])      # ✅ 29  (Molecule Man)
    print(data[1]["age"])      # ✅ 39  (Madame Uppercut)
    print(data[0].get("age"))  # ✅ 29  (get 방식)


#JSON이 { }로 시작하면 딕셔너리, [ ]로 시작하면 리스트라는 것만 기억하면 됩니다!
`````
